```python
import pandas as pd
```

```python
#Leitura dos arquivos
df1 = pd.read_excel ("./datasets/Aracaju.xlsx")
df2 = pd.read_excel ("./datasets/Fortaleza.xlsx")
df3 = pd.read_excel ("./datasets/Natal.xlsx")
df4 = pd.read_excel ("./datasets/Recife.xlsx")
df5 = pd.read_excel ("./datasets/Salvador.xlsx")
df3.head()
```

```python
#Juntandos todos os arquivos 
df = pd.concat([df1, df2, df3, df4, df5])
df.head()
```

```python
df.tail()
```

```python
#Pegando n amostras aleatórias
df.sample(6)
```

```python
df.dtypes
```

```python
#Alterando o tipo de dado da coluna X
df["LojaID"] = df["LojaID"].astype("object")
df.dtypes
```

```python
#Consultando linhas com valores faltando (somando linhas que o valor é nulo)
df.isnull().sum()
```

```python
#Substituindo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
#Substituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace = True)
```

```python
#Apagando as linhas com os valores nulos
df.dropna(inplace=True)
```

```python
#Apagando as linhas com valores nulos em apenas uma coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

```python
#Craindo novas colunas (Receitas = vendas * quantidade)
df["Receita"] = df["Vendas"].mul(df["Qtde"])
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
df.head()
```

```python
#Retorna o top N com maior valor
df.nlargest(3, "Vendas")
```

```python
#Retorna o top N com menor valor
df.nsmallest(3, "Vendas")
```

```python
#Agrupando por cidade e receiras
df.groupby("Cidade")["Vendas"].sum()
```

```python
#Ordanando conjunto de dados (Ordenando da maior quantidade de vendas para menor)
df.sort_values("Vendas", ascending=False).head(10)
```
