```python
import pandas as pd
```

```python
df1 = pd.read_excel("./datasets/Aracaju.xlsx")
df2 = pd.read_excel("./datasets/Fortaleza.xlsx")
df3 = pd.read_excel("./datasets/Natal.xlsx")
df4 = pd.read_excel("./datasets/Recife.xlsx")
df5 = pd.read_excel("./datasets/Salvador.xlsx")
```

```python
df = pd.concat([df1, df2, df3, df4, df5])
df.head()
```

```python
df.isnull().sum()
```

```python
df["Vendas"].fillna(0, inplace = True)
df.isnull().sum()
```

```python
df.head()
```

```python
df["Data"] = df["Data"].astype("int64")
df.dtypes
```

```python
#transformando a coluna Data em tipo Data
df["Data"] = pd.to_datetime(df["Data"])
df.dtypes
```

```python
#Agrupando as Vendas por ano
df.groupby(df["Data"].dt.year).head()
```

```python
#Criando uma nova coluna baseada no ano das vendas
df["Anos Vendas"] = df["Data"].dt.year
df.sample(5)
```

```python
#Extraindo o mês e o dia
df["Mês vendas"], df["Dia Vendas"] = (df["Data"].dt.month, df["Data"].dt.day)
df.sample(5)
```

```python
#Retornando a data mais antiga
df["Data"].min()
```

```python
#Retornando a data mais recente
df["Data"].max()
```

```python
#Criando diferença de dias
df["Diferença de Dias"] = df["Data"].sub(df["Data"].min())
df.sample(5)
```

```python
#Criando a coluna trimestral
df["trimestre de vendas"] = df["Data"].dt.quarter
df.sample(5)
```

```python
#Filtrando as vendas de março de 2019
vendas_marco_2019 = df.loc[(df["Data"].dt.year) & (df["Data"].dt.month == 3)]
vendas_marco_2019.sample(5)
```
